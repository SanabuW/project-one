```python
import pandas as pd 

```

```python
#using Tanlin's code
#importing both csv's of vehicles
vehicles = "all_vehicles_fix.csv"
allvehicles = pd.read_csv(vehicles, encoding= "ISO-8859-1")

evvehicles = "EV_vehicles_fix.csv"
ev = pd.read_csv(evvehicles, encoding= "ISO-8859-1")
```

```python
allvehicles
```

```python
ev
```

```python
allvehicles["is_EV"] = pd.Series(dtype = 'boolean')
allvehicles["is_EV"] = 0
allvehicles.loc[0, "is_EV"].dtype
```

```python
for index, row in allvehicles.iterrows():
    if row["Model"] in ev["Manufacturer and Model"].unique():
        print(f'{row["Model"]} is an EV at index {index}')
#        row["is_EV"] = 1 THIS DOESN'T WORK - RECORD IN NOTES
        allvehicles.loc[index, "is_EV"] = 1
allvehicles.head(50)
    



    
```

```python
allgasvehicles = allvehicles.loc[allvehicles["is_EV"] == 0]
allgasvehicles
```

```python
#borrowing Tanlin's code
allgasvehicles = allgasvehicles.dropna(subset = ["Model", "Jan","Feb","Mar","Apr","May","Jun","Jul","Aug", "Sep","Oct","Nov","Dec"])

allgasvehicles


```

```python

#borrowing Tanlin's code
allgasvehicles = allgasvehicles.replace(["-"],"0")
allgasvehicles["Jan"] = allgasvehicles["Jan"].str.replace(",","").astype(float)
allgasvehicles["Feb"] = allgasvehicles["Feb"].str.replace(",","").astype(float)
allgasvehicles["Mar"] = allgasvehicles["Mar"].str.replace(",","").astype(float)
allgasvehicles["Apr"] = allgasvehicles["Apr"].str.replace(",","").astype(float)
allgasvehicles["May"] = allgasvehicles["May"].str.replace(",","").astype(float)
allgasvehicles["Jun"] = allgasvehicles["Jun"].str.replace(",","").astype(float)
allgasvehicles["Jul"] = allgasvehicles["Jul"].str.replace(",","").astype(float)
allgasvehicles["Aug"] = allgasvehicles["Aug"].str.replace(",","").astype(float)
allgasvehicles["Sep"] = allgasvehicles["Sep"].str.replace(",","").astype(float)
allgasvehicles["Oct"] = allgasvehicles["Oct"].str.replace(",","").astype(float)
allgasvehicles["Nov"] = allgasvehicles["Nov"].str.replace(",","").astype(float)
allgasvehicles["Dec"] = allgasvehicles["Dec"].str.replace(",","").astype(float)

```

```python
allgasvehicles
```

```python
#borrowing Tanlin's code
allgasvehicles["sale total"] = allgasvehicles.sum(axis=1)
allgasvehicles
```

```python
#borrowing Tanlin's code
allgasvehicles_sorted = allgasvehicles.sort_values(["sale total"], ascending = False)  
allgasvehicles_sorted
```

```python

allgasvehicles_reIndexed = allgasvehicles_sorted.reset_index(drop = True)
allgasvehicles_reIndexed.head(10)
```

```python
allgasvehicles_reIndexed.to_csv("../resources/gas_top_sale_list.csv")
```
