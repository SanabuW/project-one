```python
originaldf.columns

# Show percentage of NaN values
# Setup removal recommendation list
removalIndicies = []

for x in range (len(originaldf.columns)):
    if (1 - (originaldf.iloc[:,x].count())/(len(originaldf))) > 0.7:
        print(f'Percentage of NaN values in series "{originaldf.iloc[:,x].name}":{1 - ((originaldf.iloc[:,x].count())/(len(originaldf)))}')
        removalIndicies.append(originaldf.iloc[:,x].name)
```

```python
removalIndicies
```

```python
# Remove the rows with excessive NaN values

colremoved = originaldf.drop(removalIndicies, axis = 1)
colremoved.head(3)
```

```python
# Ensure only electric, non-hydrogen stations are provided 

altStationsClean = colremoved.loc[colremoved["Fuel Type Code"] == "ELEC",:]
altStationsClean.head(3)
```

```python
altStationsClean.to_csv("alt_fuel_stations_cleaned.csv", index = False , header = True)
```
