```python
import pandas as pd 

vehicles = "all_vehicles_fix.csv"
allvehicles = pd.read_csv(vehicles, encoding= "ISO-8859-1")

evvehicles = "EV_vehicles_fix.csv"
    
ev = pd.read_csv(evvehicles, encoding= "ISO-8859-1")
```

```python
data = pd.merge( ev, allvehicles, left_on = "Manufacturer and Model", right_on = "Model",  how = "left") 

new_data = data[["Manufacturer and Model", "Top speed", "Acceleration","Charging time", "Nominal range", "Jan","Feb","Mar","Apr","May","Jun","Jul","Aug", "Sep","Oct","Nov","Dec"]]
new_data = new_data.dropna(subset = ["Manufacturer and Model", "Jan","Feb","Mar","Apr","May","Jun","Jul","Aug", "Sep","Oct","Nov","Dec"])

new_data = new_data.replace(["-"],"0")


new_data
```

```python
new_data.dtypes
```

```python
new_data["Jan"] = new_data["Jan"].str.replace(",","").astype(float)
new_data["Feb"] = new_data["Feb"].str.replace(",","").astype(float)
new_data["Mar"] = new_data["Mar"].str.replace(",","").astype(float)
new_data["Apr"] = new_data["Apr"].str.replace(",","").astype(float)
new_data["May"] = new_data["May"].str.replace(",","").astype(float)
new_data["Jun"] = new_data["Jun"].str.replace(",","").astype(float)
new_data["Jul"] = new_data["Jul"].str.replace(",","").astype(float)
new_data["Aug"] = new_data["Aug"].str.replace(",","").astype(float)
new_data["Sep"] = new_data["Sep"].str.replace(",","").astype(float)
new_data["Oct"] = new_data["Oct"].str.replace(",","").astype(float)
new_data["Nov"] = new_data["Nov"].str.replace(",","").astype(float)
new_data["Dec"] = new_data["Dec"].str.replace(",","").astype(float)
```

```python
# new_data["Feb"] = new_data["Feb"].str.replace(",","").astype(float)

# new_data["sale total"] = new_data[["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug", "Sep","Oct","Nov","Dec"]].map(sum)


new_data["sale total"] = new_data.sum(axis=1)
new_data = new_data[["Manufacturer and Model", "Top speed", "Acceleration","Charging time", "Nominal range", "sale total"]]
new_data
```

```python
final_data = new_data.drop_duplicates(subset="Manufacturer and Model" , keep="first", inplace=False, ignore_index=False)
final_data
```

```python
final_data = final_data.sort_values(["sale total"], ascending = False)  
final_data
```

```python
final_data = final_data.reset_index(drop = True)
final_data
```

```python
final_data.to_csv("../resources/EVtop_sale_list.csv")
```
